# Linkage

```python
import pandas as pd
import recordlinkage
import recordlinkage.preprocessing
import numpy as np
```

```python
indexer = recordlinkage.Index()
indexer.full()
candidate_links = indexer.index(new_localities,old_localities)
```

```python
new_localities['metaphone'] = recordlinkage.preprocessing.phonetic(new_localities['master_name'],method="metaphone")
old_localities['metaphone'] = recordlinkage.preprocessing.phonetic(old_localities['master_name'],method="metaphone")
```

```python
compare_booths = recordlinkage.Compare()

#Get Levenshtein score booth name
compare_booths.string('master_name', 'master_name', method='levenshtein', threshold=None, label='name_lv_score')

#Get Jaro-Winkler Score for booth name
compare_booths.string('master_name', 'master_name', method='jarowinkler', threshold=None, label='name_jw_score')

#Get Jaro-winkler score for the way the name is pronounced
compare_booths.string('metaphone', 'metaphone', method='levenshtein', threshold=None, label='metaphone')

#Get score for how far apart the booth numbers are
compare_booths.numeric('booth_rank', 'booth_rank', label='booth_number_score',method="gauss",offset=3,scale=5)
```

```python
features = compare_booths.compute(candidate_links, new_localities,old_localities)
```

```python
features
```

```python
match_index = features.index.values
master_name_index = [x[0] for x in match_index]
new_names = new_localities.iloc[master_name_index].copy()

old_name_index = [x[1] for x in match_index]
old_names = old_localities.iloc[old_name_index]
```

```python
old_names
```

```python
new_names['old_name'] = old_names.master_name.tolist()
new_names['old_booth_rank'] = old_names.booth_rank.tolist()
new_names['name_jw_score'] = features.name_jw_score.tolist()
new_names['name_lv_score'] = features.name_lv_score.tolist()
new_names['booth_score'] = features.booth_number_score.tolist()
new_names['metaphone'] = features.metaphone.tolist()
#master_names['duplicate_candidate_name'] = duplicate_names.name2.tolist()
```

```python
new_names['average_score'] = new_names[['name_jw_score','name_lv_score','booth_score']]\
                                .apply(lambda x:x.mean(),axis=1).round(2)
```

```python
new_names.columns
```

```python
new_names = new_names[['master_name','old_name','booth_rank','old_booth_rank',
                            'name_jw_score',"name_lv_score",'booth_score','average_score']]
```
